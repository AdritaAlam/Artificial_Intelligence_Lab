# Import modules

```python
import numpy as np
import matplotlib.pyplot as plt
```

#  Define constants

```python
x_val = np.arange(-10,10,0.1)

w_sig = 0.5
b_sig = -1

w_tan = -1
b_tan = 1

w_linear = 2
b_linear = 3

w1 = 0.5
b1 = -1

w2 = -1
b2 = 2

w3 = 0.5
w4 = -0.5

b = 0.5

```

# Define functions

### Function 1: Linear function
#### y = wx + b

```python
def y1(x,w,b):
  return (w*x + b)
```

### Function 2: *Quadratic* function
#### y = x^2

```python
def y2(x):
  return (x**2)
```

### Function 3: Sigmoid function
#### y = 1/(1+e^(-x))

```python

def y3(x):
  return (1/(1+np.exp(-x)))
```

### Function 4
#### y (= e^x - e^(-x)) / (e^x - e^(-x))

```python
def y4(x):
  return (np.exp(x) - np.exp(-x))/(np.exp(x) - np.exp(-x))
```

### Function 5: composite function
#### y = g(f(x)) where u = f(x) = wx + b and y = g(u) = 1/ (1 + e^(-u))

```python
def Composite_func_1(x,w,b):
  u = w*x + b
  g_of_u = 1/(1+np.exp(-u))
  return g_of_u
```

### Function 6: Composite function
#### y = g(f(x)) where f(x) = wx + b, u = f(x), and g(u) = (e^u - e^(-u)) / (e^u + e^(-u))

```python
def Composite_func_2(x,w,b):
  u = w*x + b
  g_of_u = (np.exp(u) - np.exp(-u))/(np.exp(u) + np.exp(-u))
  return g_of_u
```

### Function 7: Composite function
#### y = g3(f3(v)) where f3(v) = w3y1 + w4y2 + b, w = f3(v), and g3(w) = 1/ (1 + e^(-w))

```python
#   y1 = g1(f1(x)) where f1(x) = w1x + b1, u1 = f1(x), and g1(u1) = 1/ (1 + e^(-u1))
#   y2 = g2(f2(x)) where f2(x) = w2x + b2, u2 = f2(x), and g2(u2) = 1/ (1 + e^(-u2))
#   y = g3(f3(v)) where f3(v) = w3y1 + w4y2 + b, w = f3(v), and g3(w) = 1/ (1 + e^(-w))

def Composite_func_3(x,w1,b1,w2,b2,w3,w4,b):
  u1 = w1 * x + b1
  u2 = w2 * x + b2

  y1 = (1/(1+np.exp(-u1)))
  y2 = (1/(1+np.exp(-u2)))

  v = w3 * y1 + w4 * y2 + b
  g_of_v = 1/(1+np.exp(-v))
  return g_of_v
```

```python
y_val_1 = y1(x_val,w_linear,b_linear)  # Linear function
y_val_2 = y2(x_val)  # Quadratic function
y_val_3 = y3(x_val)  # Sigmoid function
y_val_4 = y4(x_val)
y_val_5 = Composite_func_1(x_val,w_sig,b_sig)  # Composite function
y_val_6 = Composite_func_2(x_val,w_tan,b_tan)  # Composite function
y_val_7 = Composite_func_3(x_val,w1,b1,w2,b2,w3,w4,b)  # Composite function
```

# Displaying output results

```python
print("x       y1        y2        y3        y4        y5        y6        y7")
for i in range(len(x_val)):
  print(f"{x_val[i]:.1f}    {y_val_1[i]:.2f}    {y_val_2[i]:.2f}    {y_val_3[i]:.2f}    {y_val_4[i]:.2f}    {y_val_5[i]:.2f}    {y_val_6[i]:.2f}    {y_val_7[i]:.2f}")
```

# Plotting the resulted curves

```python
plt.figure(figsize= (20,12))


plt.subplot(4,2,1)
plt.plot(x_val,y_val_1)
plt.title('Linear function')
plt.grid('on')

plt.subplot(4,2,2)
plt.plot(x_val,y_val_2)
plt.title('Quadratic function')
plt.grid('on')

plt.subplot(4,2,3)
plt.plot(x_val,y_val_3)
plt.title('Sigmoid function')
plt.grid('on')

plt.subplot(4,2,4)
plt.plot(x_val,y_val_4)
plt.title('y = (e^x - e^(-x)) / (e^x - e^(-x))')
plt.grid('on')

plt.subplot(4,2,5)
plt.plot(x_val,y_val_5)
plt.title('y = g(f(x)); f(x) = wx + b and g(u) = 1/ (1 + e^(-u))')
plt.grid('on')

plt.subplot(4,2,6)
plt.plot(x_val,y_val_6)
plt.title('Composite function:y = g(f(x))')
plt.grid('on')

plt.subplot(4,2,7)
plt.plot(x_val,y_val_7)
plt.title('Composite function:y3 = g3(f3(x))')
plt.grid('on')


plt.show()
```
