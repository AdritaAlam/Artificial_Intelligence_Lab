```python
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
def LinearFunction(x,a,b):
    return (a*x + b)
```

```python
def SineFunction(x,a,b):
    return (a*np.sin(x) + b)
```

```python
def ExponentialFunction(x,a,b):
    return (a*np.exp(x) + b)
```

```python
def QuadraticFunction(x,a,b):
    return (a*(x**2) + b)
```

```python
# a = 3
# b =  4
a = int(input('Enter the value of a: '))
b = int(input('Enter the value of b: '))

x = np.linspace(-10,10,100)

y_linear_output = LinearFunction(x,a,b)
y_sine_curve = SineFunction(x,a,b)
Y_exponential_curve = ExponentialFunction(x,a,b)
y_quadratic_curve = QuadraticFunction(x,a,b)

```

```python
plt.figure(figsize=(12,8))

plt.subplot(2,2,1)
plt.title('Linear function')
plt.plot(x,y_linear_output,label = f'y = {a}x + {b}',marker = 'o',color = 'orange',linewidth = 1)
plt.legend()

plt.subplot(2,2,2)
plt.title('Sine function')
plt.plot(x,y_sine_curve,label = f'y = {a}sin(x) + {b}',marker = 'o', color = 'orange',linewidth = 1)
plt.legend(loc = 'lower right')

plt.subplot(2,2,3)
plt.title('Exponential function')
plt.plot(x,Y_exponential_curve,label = f'y = {a}e^x + {b}',marker = 'o', color = 'orange',linewidth = 1)
plt.legend()

plt.subplot(2,2,4)
plt.title('Quadratic function')
plt.plot(x,y_quadratic_curve,label = f'y = {a}x^2 + {b}',marker = 'o', color = 'orange',linewidth = 1)
plt.legend()


plt.tight_layout()
plt.show()
```
